# Research Questions

## 1. Testing the Claim about Salaries:

```python
# Filter the data for Computer Science Engineering and specified roles
cs_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
filtered_data = df[(df['Degree'] == 'computer science & engineering') & (df['Designation'].isin(cs_roles))]

# Drop rows with missing values in the 'Salary' column
filtered_data = filtered_data.dropna(subset=['Salary'])

# Display summary statistics of salaries for this subset
print(filtered_data['Salary'].describe())

# Check the percentage of individuals earning above 2.5 lakhs
if not filtered_data.empty:
    percentage_above_threshold = (filtered_data[filtered_data['Salary'] > 2.5]['Salary'].count() / filtered_data['Salary'].count()) * 100
    print(f"Percentage of individuals earning above 2.5 lakhs: {percentage_above_threshold:.2f}%")
else:
    print("No valid data to calculate percentage.")
```

## 2. Relationship between Gender and Specialization:

```python
from scipy.stats import chi2_contingency

# Create a contingency table for Gender and Specialization
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the p-value
print("p-value:", p)
```
